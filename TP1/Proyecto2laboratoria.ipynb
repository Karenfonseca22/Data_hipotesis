```python
### TEST PARA SPEECHNESS
from scipy.stats import mannwhitneyu

# Listar categorias
categorias = ['categ_speech']

# Almancenar los datos
resultados = []

for caracteristicas in categorias:
    # Separar los datos en dos grupos según la categoría
    grupo_bajo = df[df[caracteristicas] == 'Bajo']['streams_limpio']
    grupo_medio = df[df[caracteristicas] == 'Medio']['streams_limpio']

    # Verificar que los grupos no estén vacíos antes de realizar la prueba
    if len(grupo_bajo) > 0 and len(grupo_medio) > 0:

      grupo_bajo = grupo_bajo.astype(float)
      grupo_medio = grupo_medio.astype(float)
      # Aplicar el test de Mann-Whitney U entre los grupos medio y bajo, corrected indentation
      stat_medio_bajo, p_valor_medio_bajo = mannwhitneyu(grupo_medio, grupo_bajo)

      # Almacenar los resultados
      resultados.append({
            'caracteristicas': caracteristicas,
            'stat_medio_bajo': stat_medio_bajo,
            'p_valor_medio_bajo': p_valor_medio_bajo
        })

# Mostrar los resultados
for resultado in resultados:
    print(f"Características: {resultado['caracteristicas']}")
    print(f"Estadístico de prueba Medio vs Bajo (Mann-Whitney U): {resultado['stat_medio_bajo']}")
    print(f"Valor p Medio vs Bajo: {resultado['p_valor_medio_bajo']}")
```

```python
from scipy.stats import kruskal

# Listar categorias
categorias = ['categ_dance', 'categ_valance', 'categ_energy', 'categ_acoustic', 'categ_instrumental', 'categ_liveness']

# Almacenar los resultados
resultados = []

for caracteristica in categorias:
    # Separar los datos en tres grupos según la categoría
    grupo_alto = df[df[caracteristica] == 'Alto']['streams_limpio']
    grupo_bajo = df[df[caracteristica] == 'Bajo']['streams_limpio']
    grupo_medio = df[df[caracteristica] == 'Medio']['streams_limpio']

    # Verificar que los grupos no estén vacíos antes de realizar la prueba
    if len(grupo_alto) > 0 and len(grupo_bajo) > 0 and len(grupo_medio) > 0:
        grupo_alto = grupo_alto.astype(float)
        grupo_medio = grupo_medio.astype(float)
        grupo_bajo = grupo_bajo.astype(float)

        # Aplicar el test de Kruskal-Wallis para comparar los tres grupos
        estadistico, p_valor = kruskal(grupo_alto, grupo_bajo, grupo_medio)

        # Almacenar los resultados
        resultados.append({
            'caracteristica': caracteristica,
            'estadistico_kruskal': estadistico,
            'p_valor_kruskal': p_valor,
        })

# Mostrar los resultados
for resultado in resultados:
    print(f"Característica: {resultado['caracteristica']}")
    print(f"Estadístico de prueba Kruskal-Wallis: {resultado['estadistico_kruskal']}")
    print(f"Valor p Kruskal-Wallis: {resultado['p_valor_kruskal']}")
```
